### 2. Use Euclideon Distance as a Metric for UBCF

```python
movie_data.columns
```

```python
from sklearn.metrics import pairwise_distances
```

```python
euclideon_pivot_table = pd.pivot_table(data = movie_data,values='rating',index='userId',columns='movie').fillna(0)
euclideon_pivot_table
```

```python
movie_data['userId'].unique()
```

```python
euclideon_pivot_table.index = movie_data['userId'].unique()
euclideon_pivot_table
```

```python
euclideon_pivot_table.values
```

```python
euclideon_ubcf = pairwise_distances(X =euclideon_pivot_table.values,metric='euclidean')
euclideon_ubcf
```

```python
movie_data['userId'].unique()
```

```python
ubcf_euclideon_metric         = pd.DataFrame(data = euclideon_ubcf)
ubcf_euclideon_metric.index   = movie_data['userId'].unique()
ubcf_euclideon_metric.columns = movie_data['userId'].unique()
ubcf_euclideon_metric.round(2)
```

### 3. Use Cosine as a metric for UBCF

```python
cosine_pivot_table = pd.pivot_table(data = movie_data,values='rating',index='userId',columns='movie').fillna(0)
cosine_pivot_table
```

```python
cosine_pivot_table.index = movie_data['userId'].unique()
cosine_pivot_table
```

```python
cosine_pivot_table.values
```

```python
cosine_ubcf = 1 - pairwise_distances(X = cosine_pivot_table.values,metric='cosine')
cosine_ubcf
```

```python
movie_data['userId'].unique()
```

```python
cosine_ubcf_df         = pd.DataFrame(data=cosine_ubcf)
cosine_ubcf_df.index   = movie_data['userId'].unique()
cosine_ubcf_df.columns = movie_data['userId'].unique()
cosine_ubcf_df.round(2)
```

### Filter the data top 50users
